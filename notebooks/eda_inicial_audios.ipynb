```python
import os
import sounddevice as sd

import soundfile as sf
import librosa
import librosa.display
import wave
import IPython.display as ipd

from scipy.io.wavfile import write
import matplotlib.pyplot as plt
```

## Check the sampling ratio of the audios

```python
FOLDER_PATHS = ['../dados/falas_04/']

required_sr = 8e3 # Hz
audios_without_required_sr = {}
number_of_files = 0

for path in FOLDER_PATHS: 
    for file_name in os.listdir(path):
        with wave.open(path + file_name, "rb") as wave_file:

            number_of_files += 1
            frame_rate = wave_file.getframerate()
            if frame_rate != required_sr:
                audios_without_required_sr[file_name] = frame_rate

print("Number of speech signals without the required sampling rate",
      len(audios_without_required_sr), 'of', number_of_files)
```

```python
for i, j in audios_without_required_sr.items():
    print(i, j)
```

**Basically all the speech signals must be resampled from 11.025 KHz to 8Khz**
